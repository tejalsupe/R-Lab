```r
# 4-A. find out how many values are missing in each of the columns of your data set.
values <- comma$target
sapply(comma, function(values) sum(is.na(values)))
```

```r
# 4-B. Select colum with more than 10% values missing for deletation. Create new copy of dataset that has one less column call it sample_data_1less.
# Remove columns with more than 10% NA
missing <-comma[, which(colMeans(!is.na(comma)) > 0.1)]
```

```r
# 4-C. Find 5 point summary of all remaining columns with numric attributes
numeric_only <- unlist(lapply(missing , is.numeric))
var <- missing[ , numeric_only]
summary(var)
```

```r
# 4-D. For attribute having maximum spread delete all tuples that have value less than the mean of the
# attribute selected. Count the number of tuples reamining in the sampled dataset.
mn <- mean(missing$Age, na.rm = FALSE)
mn

dd <- subset(missing , missing$Age<mn)
summary(dd$Age)
print('Rows before row deletion: ')
nrow(missing)
print('Rows before row deletion: ')
nrow(dd)
```
